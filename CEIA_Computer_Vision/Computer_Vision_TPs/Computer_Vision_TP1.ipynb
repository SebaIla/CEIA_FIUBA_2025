```python
'''
Importing required libraries for computer vision tp01
'''

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

1.Implementar el algoritmo de White Patch para librarnos de las diferencias de color de iluminación.

```python
# Definición de la función white_patch

def white_patch(image, percentile=100):
    # Conversión de la imagen a tipo float32 para calculos de maximos y percentiles
    img_float = image.astype(np.float32)
    
    # Encontrar valor máximo o percentil para cada canal
    if percentile == 100:
        ref = np.max(img_float, axis=(0, 1))
    else:
        ref = np.percentile(img_float, percentile, axis=(0, 1))
    
    # Asegurar que el valor de referencia no sea menor a 1 para evitar división por cero
    ref = np.clip(ref, 1, None)
    
    # Corrección de la imagen
    corrected = img_float * (255.0 / ref)
    corrected = np.clip(corrected, 0, 255).astype(np.uint8)
    return corrected

image_1 = cv.imread('test_blue.png')
image_2 = cv.imread('test_red.png')
image_3 = cv.imread('test_green.png')

balanced_image_1 = white_patch(image_1, percentile=90)
balanced_image_2 = white_patch(image_2, percentile=90)
balanced_image_3 = white_patch(image_3, percentile=90)

cv.imwrite('balanced_test_blue.png', balanced_image_1)
cv.imwrite('balanced_test_red.png', balanced_image_2)
cv.imwrite('balanced_test_green.png', balanced_image_3)
```
